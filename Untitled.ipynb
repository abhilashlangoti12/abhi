```python
n=input("enter the number: ")
n1=input("enter the 1st number to replace:")
n2=input("enter the 2nd numvber to replace:")
num_list=list(n)
print("num list",num_list)
for i in range(len(num_list)):
    if num_list[i]==n1:
        num_list[i]=n2
    elif num_list[i]==n2:
        num_list[i]=n1
print("replaced number list:",num_list)
n="".join(num_list)
print(n)
```

```python
lst=list(input("enter the list of strigs ").split())
result_lst = []
for word in lst:
    up_word = word.capitalize()
    result_lst.append(up_word)
print("updated list:",result_lst)
```

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
occ = {}
for n in numbers:
    if n in occ:
        occ[n] += 1
    else:
        occ[n] = 1
for key, value in occ.items():
    print(key, ":", value)
```

```python
lst = list(map(int, input("Enter numbers separated by space: ").split()))

freq = {}

for n in lst:
    if n in freq:
        freq[n] += 1
    else:
        freq[n] = 1

print("Frequency of elements:")
for k, v in freq.items():
    print(k, ":", v)
```

```python
d = {'a': 10, 'b': 25, 'c': 5, 'd': 30}

max_key = max(d, key=d.get)

print("Key with maximum value:", max_key)
print("Value:", d[max_key])
```

```python
d = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}

unique = {}
seen = set()

for k, v in d.items():
    if v not in seen:
        unique[k] = v
        seen.add(v)

print("Dictionary after removing duplicate values:", unique)
```
